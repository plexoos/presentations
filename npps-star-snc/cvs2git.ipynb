```python
import os
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
import collections
```

```python
values = [
    ('R0', 392, 275, 992),
    ('R1', 383, 261, 992),
    ('R2', 340, 252, 911),
    ('R3', 262, 174, 449),
    ('R4', 259, 171, 419),
    ('R5', 254, 165, 384),
]
```

```python
df_sizes = pd.DataFrame(values, columns=['name', 'bare1', 'bare2', 'co'])
df_sizes
```

```python
df_sizes_norm = df_sizes.melt(id_vars='name')#, value_vars=['size', 'sss'])
df_sizes_norm
```

```python
fig = plt.figure(figsize=(6,4))
ax = sns.barplot(x='name', y='value', hue='variable', data=df_sizes_norm)
ax.legend(ax.legend().get_patches(), ['.git (step 1)', '.git (step 2)', '.git (step 2) + checkout'])
ax.set(xlabel='Scenario', ylabel='Size, MB')
#ax.semilogy();
ax.grid()
plt.tight_layout()
plt.savefig('graphics/cvs2git.png')
```
