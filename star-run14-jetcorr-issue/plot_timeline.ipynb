```python
import matplotlib
import matplotlib.pyplot as plt


with plt.xkcd():
    font = {'size': 20}
    matplotlib.rc('font', **font)
    
    fig, ax = plt.subplots(1, 1, figsize=(15, 5), dpi=100)
    
    ax.set_ylim([-15, 5])
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.yaxis.set_ticks_position('none')
    
    y = [0, 0,   0,   0,   -10, -10,
         -10, -10, -10,
         -10, -10, -10, 0, 0, 0, 0]
    x = [-0.3, 0, 0.7,   1,   1, 1.3,
         2.2, 2.5, 2.8,
         3.3, 4.5,   5, 5, 5.5, 5.8, 6]
    labels = ['', 'SL16d', 'SL16f', '', '', 'SL16g',
              'SL17a', '', 'SL17b',
              'SL17d', 'SL18f', '', '', '', 'SL18h-rc', '']
    
    y_sim = [-10,     -5, -5, -10]
    x_sim = [  2.5,  2.5,  4.8, 4.8]
    
    ax.set_xticks(x)
    ax.set_yticks([])
    ax.set_xticklabels(labels)
    
    xticks = ax.xaxis.get_major_ticks()
    xticks[4].set_pad(-55)
    xticks[7].set_pad(-55)
    
    xticks[0].set_visible(False)
    xticks[3].set_visible(False)
    xticks[-1].set_visible(False)
    xticks[-3].set_visible(False)
    xticks[-4].set_visible(False)
    xticks[-5].set_visible(False)
    
    
    plt.xlabel('Timeline', labelpad=5)
    plt.ylabel('Global Tracks')
    
    plt.annotate('Simulation only', xy=(3.5, -4.5), xytext=(-80, 50), textcoords='offset points',
                 arrowprops=dict(arrowstyle='fancy'))
    
    plt.annotate('~10% drop', xy=(1.02, -5), xytext=(30, 10), textcoords='offset points',
                 arrowprops=dict(arrowstyle='fancy'))
    plt.annotate('~5% gain', xy=(2.48, -8), xytext=(-90, 20), textcoords='offset points',
                 arrowprops=dict(arrowstyle='fancy'))
    plt.annotate('Recover', xy=(4.78, -8), xytext=(-100, 10), textcoords='offset points',
                 arrowprops=dict(arrowstyle='fancy'))
    plt.annotate('Recover', xy=(5.02, -3.5), xytext=(30, 20), textcoords='offset points',
                 arrowprops=dict(arrowstyle='fancy'))
    
    plt.annotate('Issue #1\n2016-06-29', xy=(1, -10.2), xytext=(-50, -70), textcoords='offset points',
                 arrowprops=dict(arrowstyle='->'))
    plt.annotate('Issue #2\n2017-02-14', xy=(2.5, -10.2), xytext=(-50, -70), textcoords='offset points',
                 arrowprops=dict(arrowstyle='->'))

    plt.ylim(-18, 2)
    plt.plot(x, y, 'C0-', x_sim, y_sim, 'C0--')


plt.tight_layout()
plt.savefig('timeline.png')
plt.show()
```
